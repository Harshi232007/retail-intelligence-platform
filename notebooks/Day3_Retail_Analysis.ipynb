```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("C:\\Users\\Harshitha\\OneDrive\\Desktop\\Retail-Intelligence-Platform\\data\\cleaned_superstore.csv")
```

```python
total_sales=df["Sales"].sum()
print(total_sales)
```

```python
total_profit=df["Profit"].sum()
print(total_profit)
```

```python
category_sales=df.groupby("Category")["Sales"].sum()
print(category_sales)
```

```python
category_profit=df.groupby("Category")["Profit"].sum()
print(category_profit)
```

```python
top_products=df.groupby("Product.Name")["Sales"].sum().sort_values(ascending=False).head(10)
print(top_products)
```

```python
top_customer=df.groupby("Customer.Name")["Sales"].sum().sort_values(ascending=False).head(10)
print(top_customer)
```

```python
state_sales=df.groupby("State")["Sales"].sum().sort_values(ascending=False).head(10)
print(state_sales)
```

```python
average_discount=df["Discount"].mean()
print(average_discount)
```

```python
category_sales.plot(kind="bar")
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

```python
category_profit.plot(kind="bar")
plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.show()
```

```python
top_products.plot(kind="bar")
plt.title("Top 10 products")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()
```
